```python
from PIL import Image, ImageDraw, ImageFont

# High resolution for clarity
W, H = 1200, 2000
bg_color = "#002b1c"  # Deep Forest Green
card_color = "#ffffff"
gold = "#d4af37"
text_dark = "#001a11"

img = Image.new("RGB", (W, H), bg_color)
draw = ImageDraw.Draw(img)

# Use default font (this will work on all systems)
try:
    # Try to load a larger default font
    font_bold = ImageFont.load_default()
    font_semi = ImageFont.load_default()
    font_reg = ImageFont.load_default()
    font_small = ImageFont.load_default()
    
    # For default font, we need to scale it up
    # Create a new image with text and scale it
    def draw_scaled_text(position, text, fill, font, scale=1):
        # This is a workaround for default font size limitations
        temp_img = Image.new("RGB", (W, H), bg_color)
        temp_draw = ImageDraw.Draw(temp_img)
        temp_draw.text(position, text, fill=fill, font=font)
        # Scale if needed
        if scale > 1:
            new_size = (int(W * scale), int(H * scale))
            temp_img = temp_img.resize(new_size, Image.Resampling.LANCZOS)
            img.paste(temp_img, (0, 0))
        else:
            img.paste(temp_img, (0, 0))
    
except Exception as e:
    print(f"Font error: {e}")
    # Create simple text without custom fonts
    draw.text((W/2, 120), "RAMADAN 2026", fill=gold, anchor="mm")
    draw.text((W/2, 190), "THE 30-DAY SPIRITUAL ROADMAP", fill="white", anchor="mm")

# Header Section (using simple drawing)
draw.text((W/2, 120), "RAMADAN 2026", fill=gold, anchor="mm")
draw.text((W/2, 190), "THE 30-DAY SPIRITUAL ROADMAP", fill="white", anchor="mm")
draw.line([(200, 220), (1000, 220)], fill=gold, width=3)

# Content Data
content = [
    ("Phase 1: Mercy (Days 1-10)", "Focus on compassion, patience, and kindness."),
    ("Phase 2: Forgiveness (Days 11-20)", "Seeking Istighfar and purifying the heart."),
    ("Phase 3: Salvation (Days 21-30)", "Searching for Laylatul Qadr & extra worship."),
    ("1. Five Daily Prayers", "Maintain Salah on time with Khushu (focus)."),
    ("2. Quran Recitation", "Aim to complete or read a portion daily."),
    ("3. Suhoor & Iftar", "Follow the Sunnah; keep meals light and healthy."),
    ("4. Taraweeh & Tahajjud", "Nightly prayers for spiritual elevation."),
    ("5. Charity (Sadaqah)", "Give generously to those in need."),
    ("6. Guarding the Tongue", "Avoid gossip, lying, and idle talk."),
    ("7. Daily Dhikr", "Keep your tongue moist with the remembrance of Allah.")
]

y_start = 280
card_h = 140
gap = 30

for title, desc in content:
    # Draw card
    draw.rounded_rectangle([100, y_start, 1100, y_start + card_h], radius=20, fill=card_color)
    
    # Text in card - using simpler text drawing
    draw.text((150, y_start + 35), title, fill=text_dark)
    draw.text((150, y_start + 85), desc, fill="#444444")
    
    # Gold accent bar on the left of each card
    draw.rectangle([100, y_start + 20, 115, y_start + card_h - 20], fill=gold)
    
    y_start += card_h + gap

# Footer
draw.text((W/2, H - 100), "May Allah accept our fasting and prayers.", fill="white", anchor="mm")

# Save the image
img.save("ramadan_guide_simple.png")
print("Image created successfully!")
```
